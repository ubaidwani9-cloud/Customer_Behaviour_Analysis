```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df= df.rename(columns = {'purchase_amount_(usd)' : 'purchase_amount'})
```

```python
df.columns
```

Feature Engineering:

```python
## Create new column age_groups
labels = ['Young_Adults','Adults','Middle-aged','senior']
df['age_group'] = pd.qcut(df['age'], q=4 , labels = labels)
```

```python
df[['age','age_group']].head(10)
```

```python
## Create purchase_frequency_days  Column

frequency_mapping1 = {
    'Fortnightly' : 14,
    'Weekly' : 7,
    'Monthly' : 30,
    'Quarterly' : 90,
    'Bi-Weekly' : 14,
    'Annually' : 360,
    'Every 3 Months' : 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping1)
```

```python
df[['frequency_of_purchases','purchase_frequency_days']].head(10)
```
